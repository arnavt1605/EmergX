```python
df= pd.read_csv("/kaggle/input/scores-new/scores_new.csv")
```

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
def sbert_score(reference, answer):
    if not answer.strip():
        return 0.0

    ref= model.encode(reference, convert_to_tensor=True)
    ans= model.encode(answer, convert_to_tensor=True)
    similarity= util.cos_sim(ref,ans).item()  #cosine similarity
    score= max(0.0, similarity)*100

    return score
```

```python
scores=[]
for _, row in df.iterrows():
    qid = row["question_id"]
    ans = row["answer"]
    ref = reference_answers[qid]

    score= sbert_score(ref, ans)
    scores.append(score)
```

```python
df["sbert_score"]= scores
```

```python
df = df.drop(columns=["Unnamed: 0"])
```
